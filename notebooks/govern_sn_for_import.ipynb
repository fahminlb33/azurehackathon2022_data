```python
import uuid
from faker import Faker

import numpy as np
import pandas as pd
```

```python
faker = Faker()
```

```python
df_nama = pd.read_csv("../processed/sn_nama_all.csv")
```

```python
siswa_id_to_uuid = {siswa_id: str(uuid.uuid4()) for siswa_id in df_nama.siswa_id.values}
```

```python
def map_group(x):
    if "MIPA" in x:
        return "MIPA"
    elif "IPS" in x:
        return "IPS"
    else:
        return "Bahasa"

df_nama["Group"] = df_nama.kelas.map(map_group)
df_nama["StudentId"] = df_nama.siswa_id.map(siswa_id_to_uuid)
df_nama["Name"] = df_nama.nama.map(lambda x: faker.name())
df_nama[["StudentId", "Name", "Group"]].to_csv("../processed/sn_students_import.csv", index=False)
```

```python
df_nilai = pd.read_csv("../processed/sn_nilai_all.csv")
```

```python
mapel_to_grade_dict = {
    "MTK": "Mathematics",
    "IND": "BahasaIndonesia",
    "ING": "EnglishLanguage",
    "PAI": "Religion",
    "PJK": "PhysicalEducation",
    "PKN": "CivicEducation",
    "PKR": "Crafts",
    "SBK": "Arts",
    "SUN": "BahasaSunda",
    "SI": "IndonesianHistory",


    "MTK_P": "AdvanceMathematics",
    "BIO": "Biology",
    "FIS": "Physics",
    "KIM": "Chemistry",


    "EKO": "Economy",
    "SEJ": "History",
    "GEO": "Geography",
    "SOS": "Sociology",

    "ANTR": "Anthropology",
    "SIND": "IndonesianLiterature",
    "SING": "EnglishLiterature",
    "JER": "GermanLanguage",
    "JAP": "JapaneseLanguage",
}
```

```python
df_nilai["GradeId"] = df_nilai.siswa_id.apply(lambda x: str(uuid.uuid4()))
df_nilai["StudentId"] = df_nilai.siswa_id.map(siswa_id_to_uuid)
df_nilai["Subject"] = df_nilai.mapel.map(mapel_to_grade_dict)
df_nilai = df_nilai.rename(columns={"semester": "Semester", "nilai": "Score"})
df_nilai_clean = df_nilai[["GradeId", "StudentId", "Subject", "Semester", "Score"]]
df_nilai_clean = df_nilai_clean.fillna(0)
df_nilai_clean.to_csv("../processed/sn_grades_import.csv", index=False)
df_nilai_clean
```
