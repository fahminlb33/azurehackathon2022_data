```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("../processed/sn_nilai_all.csv")
df.head()
```

```python
df_single = df[df["siswa_id"] == 1]
df_single.pivot_table(values="nilai", index="mapel", columns=["semester"])
```

```python
grades_group_dict = {
    "stem": ["MTK", "BIO", "FIS", "KIM", "MTK_P", "GEO"],
    "language": ["IND", "ING", "SIND", "SING", "JER", "ANTR", "SUN"],
    "business": ["MTK", "MTK_P", "EKO", "IND", "ING"],
    "arts": ["SBK", "PKR"],
    "social": ["SOS", "SEJ", "SI"],
    "others": ["PKN", "PJK"]
}
```

```python
# summarize by grades category
df_grades = df.pivot_table(values="nilai", index=["semester", "siswa_id"], columns="mapel")

# reset index
df_grades = df_grades.reset_index()

# group by grades category
df_grades = df_grades.assign(
    stem=lambda x: x[grades_group_dict["stem"]].median(axis=1),
    language=lambda x: x[grades_group_dict["language"]].median(axis=1),
    business=lambda x: x[grades_group_dict["business"]].median(axis=1),
    arts=lambda x: x[grades_group_dict["arts"]].median(axis=1),
    social=lambda x: x[grades_group_dict["social"]].median(axis=1),
    others=lambda x: x[grades_group_dict["others"]].median(axis=1),
)

# get only the grouped grades
df_grades = df_grades[["siswa_id", "semester", "stem", "language", "business", "arts", "social", "others"]]

# unpivot the dataframe
df_grades = df_grades.melt(id_vars=["siswa_id", "semester"])

# rename mapel to category
df_grades = df_grades.rename(columns={"mapel": "kategori"})

# save to file
df_grades.to_csv("../processed/sn_nilai_by_category.csv", index=False)

df_grades
```
